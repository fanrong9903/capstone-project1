```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
filenames = glob.glob('data/*.csv')
```

```python
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
atp_data = pd.concat(dfs, ignore_index=True)
```

```python
atp_data.shape
```

```python
atp_data.head()
```

```python
atp_data.info()
```

```python
atp_data.isnull().sum()
```

```python
# drop the columns that doesn't have  a lot of values
atp_data.drop(columns=['draw_size','winner_seed','winner_entry','loser_seed','loser_entry'],inplace=True)
```

```python
atp_data.shape
```

```python
atp_data.isnull().sum()
```

```python
atp_data.dropna(subset=['w_1stWon','loser_rank','winner_rank','l_1stWon'],inplace=True)
```

```python
atp_data.shape
```

```python
atp_data.isnull().sum()
```

```python
atp_data.winner_ht=atp_data.winner_ht.fillna(atp_data.winner_ht.median())
atp_data.loser_ht=atp_data.loser_ht.fillna(atp_data.loser_ht.median())
atp_data.loser_age=atp_data.loser_age.fillna(atp_data.loser_age.median())
atp_data.minutes=atp_data.minutes.fillna(atp_data.minutes.median())
atp_data.loser_hand=atp_data.loser_hand.fillna('U')
atp_data.winner_hand=atp_data.winner_hand.fillna('U')
```

```python
# check agian, and there are no more missing data anymore.
atp_data.isnull().sum()
```

```python
atp_data.describe()
```

```python
atp_data.tourney_date=pd.to_datetime(atp_data.tourney_date.apply(str))
```

```python
atp_data.set_index('tourney_date',inplace = True)
```
